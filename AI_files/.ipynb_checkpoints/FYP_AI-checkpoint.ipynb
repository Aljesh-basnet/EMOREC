<h2>Preprocessing and data cleaning</h2>

```python
%%time
import re 
# The motivation for the code was from a youtube video and is used to clean the datasets
text = ' '.join(df_train.iloc[:,0])
text =text.split()
common = pd.Series(text).value_counts()
rare =common[common.values == 1]

def clean(x):
    if type(x) is str:
        x=x.lower()
        x = re.sub(r'([a-zA-Z0-9+._-]+@[a-zA-Z0-9._-]+\.[a-zA-Z0-9_-]+)', '', x) 
        #regex to remove to emails
        x = re.sub(r'(http|ftp|https)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '', x)
        #regex to remove URLs
        x = re.sub('[^A-Z a-z]+', '', x)
        x = ' '.join([t for t in x.split() if t not in rare])
        return x
    else:
        return x
    
df_train.iloc[:,0] = df_train.iloc[:,0].apply(lambda x: clean(x)) 
df_test.iloc[:,0] = df_test.iloc[:,0].apply(lambda x: clean(x))
df_val.iloc[:,0] = df_val.iloc[:,0].apply(lambda x: clean(x))
```

```python
x_train =df_train.iloc[:,0].tolist()

# x =df_train.iloc[:,0].tolist()
# x[:10]
x_val=df_val.iloc[:,0].tolist()
x_test = df_test.iloc[:,0].tolist()
x_train[:10]
```

```python
df_train.iloc[:,1]=df_train.Emotions.replace({'joy':0,'anger':1,'love':2,'sadness':3,'fear':4,'surprise':5})
train_y = df_train.iloc[:,1].values

df_val.iloc[:,1]=df_val.Emotions.replace({'joy':0,'anger':1,'love':2,'sadness':3,'fear':4,'surprise':5})
val_y = df_val.iloc[:,1].values

df_test.iloc[:,1]=df_test.Emotions.replace({'joy':0,'anger':1,'love':2,'sadness':3,'fear':4,'surprise':5})
test_y = df_test.iloc[:,1].values
val_y
```

```python
x_train,y_train = np.array(x_train),np.array(train_y)
x_test,y_test = np.array(x_test),np.array(test_y)
x_val,y_val = np.array(x_val),np.array(val_y)
```

```python
# The motivation for the code was taken from a kaggle notebook related to preprocessing and is used to remove the stop words
stopwords = stopwords.words('english')
def remove_stop_words(corpus):
    removed_stop_words = []
    for review in corpus:
        removed_stop_words.append(
            ' '.join([word for word in review.split() 
                      if word not in stopwords])
        )
    return removed_stop_words

x_train = remove_stop_words(x_train)
x_test = remove_stop_words(x_test)
x_val= remove_stop_words(x_val)
```

```python
print(x_train)
```

```python
# The code was extracted from a kaggle notebook related to preprocessing is used to get the stemmed text from the dataset
def get_stemmed_text(x):
    stemmer = PorterStemmer()
    return [' '.join([stemmer.stem(word) for word in review.split()]) for review in x]

x_train = get_stemmed_text(x_train)
x_test = get_stemmed_text(x_test)
x_val = get_stemmed_text(x_val)
```

```python
print(x_train)
# print(x_test[0])
# print(x_val[0])
```

```python
# The code was extracted from a kaggle notebook related to preprocessing is used to get the lemmatized text from the dataset
def get_lemmatized_text(x):
    lemmatizer = WordNetLemmatizer()
    return [' '.join([lemmatizer.lemmatize(word) for word in review.split()]) for review in x]

x_train = get_lemmatized_text(x_train)
x_test = get_lemmatized_text(x_test)
x_val = get_lemmatized_text(x_val)
```

```python
print(x_train)
```

```python
total_text = np.concatenate((x_train,x_test,x_val),axis=0)
```

```python
# the motivation of this code was taken from tokenizer documentation and is used to provide tokens to the texts.

token =Tokenizer(lower=True,filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n',oov_token='UNK')
token.fit_on_texts(total_text)
max_length =25
def tokenize(x):
        encoded_text = token.texts_to_sequences(x)
        padded_text=pad_sequences(encoded_text,maxlen=max_length,padding='post')
        return padded_text
```

```python
# the motivation of this code was taken from the LSTM book written by Jason Brownlee and is used to one hot encode the labels
label_encoder = LabelEncoder()
onehot_encoder = OneHotEncoder(sparse=False)
def one_hot_encoding(x):
    integer_encoded = label_encoder.fit_transform(x)
    integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
    onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
    return onehot_encoded
```

```python
# The motivation for this code comes from a embedding notebook of kaggle and is used to load the globe embeddings file
embedding_index=dict()
file=open('glove.6B.100d.txt',encoding='utf8')
for line in file:
    values=line.split()
    word=values[0]
    coefs=np.asarray(values[1:],dtype='float32')
    embedding_index[word]=coefs
file.close()
print('Loaded %s word vectors '%len(embedding_index))
```

```python
# The motivation for this code comes from a embedding notebook of kaggle and is used to create the embedding matrix
num_words = len(token.word_index) +1
embedding_dim = 100

# first create a matrix of zeros, this is our embedding matrix
embedding_matrix = np.zeros((num_words, embedding_dim))

# for each word in out tokenizer lets try to find that work in our w2v model
for word, i in token.word_index.items():
    if i > num_words:
        continue
    embedding_vector = embedding_index.get(word)
    if embedding_vector is not None:
        # we found the word - add that words vector to the matrix
        embedding_matrix[i] = embedding_vector
    else:
        # doesn't exist, assign a random vector
        embedding_matrix[i] = np.random.randn(embedding_dim)
```

```python
# The motivation for this code comes from the book of LSTM written by Jason Brownlee and is used to create the LSTM model
model = Sequential()
model.add(Embedding(input_dim=num_words,output_dim =embedding_dim,weights=[embedding_matrix], input_length=max_length,trainable=True))
model.add(Dropout(0.5))
model.add(LSTM(100,return_sequences=True))
model.add(Dropout(0.5))
model.add(LSTM(50,return_sequences=True))
model.add(Dropout(0.5))
model.add(LSTM(25))
model.add(Dropout(0.5))
model.add(Dense(6,activation='softmax'))
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
# The motivation for this code comes from the book of LSTM written by Jason Brownlee and is used to train the LSTM model.
data= model.fit(X_train,Y_train,batch_size=32,epochs=10,validation_data=(X_val,Y_val))
```

```python
# The motivation for this code comes from the notebook of LSTM and is used to plot the training accuracy
# and validation accuracy of LSTM model
plt.plot(data.history['accuracy'],c='c',label='train_data')
plt.plot(data.history['val_accuracy'],c='m',label='validation')
plt.legend(loc='lower right')
plt.show()
```
